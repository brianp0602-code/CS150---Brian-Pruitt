*Task 1: Input Positive Integer and Factors*

```python

num = int(input("Please input a positive integer: "))

for i in range(1, num + 1):
    if num % i == 0:
        print(f"{i} is a factor of {num}")
```

```python

num = int(input("Please input a positive integer: "))

factor_count = 0

for i in range(1, num + 1):
    if num % i == 0:
        factor_count += 1

if factor_count == 2:
    print("Your number is prime")
else:
    print("Your number is not prime")
```

*Task 3: Temperature Summaries*

```python
louisville_temps = {
    "january":   [43, 27],
    "february":  [49, 30],
    "march":     [58, 39],
    "april":     [70, 49],
    "may":       [78, 59],
    "june":      [86, 67],
    "july":      [89, 71],
    "august":    [88, 70],
    "september": [82, 62],
    "october":   [71, 51],
    "november":  [58, 39],
    "december":  [47, 32]
}

while True:
    month = input(
        "Enter a month: I will give you the average high and low temperatures in Louisville\n: "
    ).lower()

    if month == "quit":
        break

    if month in louisville_temps:
        high, low = louisville_temps[month]
        print( f"The average high temperature in Louisville for {month.capitalize()} "
            f"is {high} and the average low is {low}."  )
    else:
        print("Not a valid month")
```

*Task 4: Inventory Dictionary and Prices *

```python

inventory = {
    "hammer": [15, 12.49],
    "wrench": [20, 7.99],
    "screwdriver": [40, 3.99],
    "nailgun": [10, 36.59],
    "padlock": [25, 5.99],
    "roll_of_duct_tape": [30, 4.95]
}

def lookup_item():
    item = input("Enter item name: ").lower()
    if item in inventory:
        qty, price = inventory[item]
        print(f"{item}: {qty} available at ${price:.2f} each")
    else:
        print("Item not available.")

def items_under_budget():
    budget = float(input("Enter maximum budget: "))
    for item, (qty, price) in inventory.items():
        if price < budget:
            print(f"{item} — ${price:.2f}")

def total_quantity():
    print(sum(qty for qty, _ in inventory.values()))

def total_value():
    total = sum(qty * price for qty, price in inventory.values())
    print(f"${total:.2f}")

while True:
    print("\n1) Lookup item")
    print("2) Items under budget")
    print("3) Total quantity")
    print("4) Total inventory value")
    print("5) Quit")

    choice = input("Choice: ")

    if choice == "1":
        lookup_item()
    elif choice == "2":
        items_under_budget()
    elif choice == "3":
        total_quantity()
    elif choice == "4":
        total_value()
    elif choice == "5":
        break
    else:
        print("Invalid option")
```
