```python
# Triangle Type Checker

a = float(input("Enter side 1: "))
b = float(input("Enter side 2: "))
c = float(input("Enter side 3: "))

if a == b == c:
    print("Equilateral triangle")
elif a == b or b == c or a == c:
    print("Isosceles triangle")
else:
    print("Scalene triangle")
```

```python
student_grades= [89,60,90,72,75]
class_average = sum(student_grades) / len(student_grades)
minimum_grade = min(student_grades)
maximum_grade = max(student_grades)

print("Class average:", class_average)
print("Minimum grade:", minimum_grade)
print("Maximum grade:", maximum_grade)
at_risk_count = sum(1 for g in student_grades if g < 65)
print(f"{at_risk_count} students are at risk in the course.")
```

```python
# Task 3: Ranges and Multiples

# A. Multiples of 2
mult_of_2 = [n for n in range(1, 101) if n % 2 == 0]

# B. Multiples of 3
mult_of_3 = [n for n in range(1, 101) if n % 3 == 0]

# C. Multiples of both (divisible by 6)
mult_of_both = [n for n in range(1, 101) if n % 6 == 0]

# D. Multiples of either (union)
mult_of_either = sorted(list(set(mult_of_2) | set(mult_of_3)))

print("Multiples of 2:", mult_of_2)
print("Multiples of 3:", mult_of_3)
print("Multiples of both:", mult_of_both)
print("Multiples of either:", mult_of_either)
```

```python
# Task 4: Shopping List Program

shopping_list = []

# A. Input up to 7 items
for i in range(7):
    item = input("Enter an item (or type 'done' to finish): ").strip()
    if item.lower() == "done":
        break
    shopping_list.append(item)

# B. Number of items
print("Total items:", len(shopping_list))

# C. Reverse order
print("Reverse order:", shopping_list[::-1])

# D. Alphabetical order
print("Alphabetical order:", sorted(shopping_list))

# E. Check for duplicates using embedded loops
duplicate_found = False

for i in range(len(shopping_list)):
    for j in range(i + 1, len(shopping_list)):
        if shopping_list[i].lower() == shopping_list[j].lower():
            duplicate_found = True
            break
    if duplicate_found:
        break

if duplicate_found:
    print("At least one item appears more than once in the list.")
else:
    print("No duplicates found.")
```
