## *Weekend activities list*

```python
# Task 1 (short version)

actSat = set()
while (a := input('Saturday activity ("Done" to stop): ')) != "Done":
    actSat.add(a)

actSun = set()
while (a := input('Sunday activity ("Done" to stop): ')) != "Done":
    actSun.add(a)

print("\nSaturday:", actSat)
print("Sunday:", actSun)
print("Both days:", actSat & actSun)
print("At least one day:", actSat | actSun)
```

## *Triangle Calculations*

```python
import math

def triangle_solve(leg1, leg2):
    hypotenuse = math.sqrt(leg1**2 + leg2**2)
    perimeter = leg1 + leg2 + hypotenuse
    area = (leg1 * leg2) / 2
    return (hypotenuse, perimeter, area)

result = triangle_solve(3, 4)
print("\nTriangle test (3, 4):", result)
```

## *Calculator Class*

```python
class Calculator:
    def __init__(self):
        self.prev_result = None
        self.total_calculations = 0

    def add(self, a, b):
        self.prev_result = a + b
        self.total_calculations += 1
        return self.prev_result

    def subtract(self, a, b):
        self.prev_result = a - b
        self.total_calculations += 1
        return self.prev_result

    def multiply(self, a, b):
        self.prev_result = a * b
        self.total_calculations += 1
        return self.prev_result

    def divide(self, a, b):
        self.prev_result = a / b
        self.total_calculations += 1
        return self.prev_result

    def power(self, a, b):
        self.prev_result = a ** b
        self.total_calculations += 1
        return self.prev_result

    def info(self):
        print("Previous result:", self.prev_result)
        print("Total calculations:", self.total_calculations)

calc = Calculator()
calc.add(5, 3)
calc.multiply(2, 4)
calc.power(2, 3)
calc.info()
```
